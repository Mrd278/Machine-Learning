```
import numpy as np
import matplotlib.pyplot as plt


class PlotFunction:
  
  def __init__(self):
    self.X = []
    self.Y = []
    
  def Enter_the_details(self):
    x = int(input("Enter x: "))
    y = int(input("Enter y: "))
    
    self.X.append(x)
    self.Y.append(y)
    
  def plotting(self):
    plt.scatter(self.X,self.Y,'r+')
```

```
p = PlotFunction()
p.Enter_the_details()
```

```
p.Enter_the_details()
p.Enter_the_details()
p.Enter_the_details()
p.Enter_the_details()
p.Enter_the_details()
```

```
p.plotting()
```
